# 1. Tuples 

Tuples are used to store multiple items in a single variable. It is a collection which is ordered and unchangeable. Tuples are written with round brackets ().

They allow duplicate values. 
Items in a tuple have a definite order, and that order will not change.
Tuple items are indexed.

** Tuples are immutable i.e unchangeable. Therefore, we cannot change, add or remove items after the tuple has been created.

```python
mytuple=('apple','banana','cherry','banana','mango','strawberry')
print(mytuple)
print(type(mytuple))
```

```python
# len(tuple_name) - prints the number of items in a tuple
print(len(mytuple))
```

### Tuple() Constructor

```python
# It is also possible to use the tuple() constructor to make a tuple.

tuple1 = tuple(('car','bike','train','bus'))
tuple3 = (True,'raman','taman',2,'shyam',89101)
print(type(tuple1))
print(type(tuple3))
```

```python
# To make a tuple with only 1 value in it, we do the following :
tuple2 = ('airways',)
print(type(tuple2))

tuple_ex = ('airways')
print(type(tuple_ex))
# return string...tuple is not formed
```

### Accessing Tuple Elements

```python
tuple4 = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango", "pineapple", "papaya")

# start from 0 as first index position
print(tuple4[2:6])
print(tuple4[:5])
print(tuple4[6:])

# negative indexes
# -1 represents last index
print(tuple4[:-1])
print(tuple4[-6:-1])
```

```python
# Check if item exists
fruit = input("Enter a fruit: ");
if fruit in tuple4:
    print('Present')
else: 
    print('Not present')
```

### Change Tuple Values

To change values in tuple, we have to convert tuple into a list, change the list, and convert the list back to tuple.

```python
# tuple1 = ('car','bike','train','bus')
tempList = list(tuple1)
tempList[2] = 'airways'
tuple1 = tuple(tempList)
print(tuple1)
```

```python
# Add items using append() for list by converting a tuple into list and updating the list and converting it back into the tuple.

tempList = list(tuple1)
tempList.append('bicycle')
tuple1 = tuple(tempList)
print(tuple1)
```

```python
# Add tuple to a tuple
tuple2 += tuple1
print(tuple2)
```

```python
# remove elements through list
tempList = list(tuple1)
tempList.remove('bicycle')
tuple1 = tuple(tempList)
print(tuple1)
```

### Unpacking Tuple

```python
letters = ('a','b','c','d','e','f','g')

# basically destructuring the tuple
(first, second, third, fourth, fifth, sixth, seventh) = letters
print(first)
print(second)
print(third)
print(fourth)
print(fifth)
print(sixth)
print(seventh)
```

```python
# The number of variables must match the number of values in the tuple, if not, you must use an asterisk to collect the remaining values as a list.
(val1, val2, val3, val4, *val5) = letters
print(val1)
print(val2)
print(val3)
print(val4)
print(val5)
```

```python
(x1, x2, *x3, x4) = letters
print(x2)
print(x3)
print(x4)
```

### Looping through tuples

```python
# For Loop
# tuple3 = (True,'raman','taman',2,'shyam',89101)
for i in range(len(tuple3)):
    print(tuple3[i], end=" ")
```

```python
# While Loop
# tuple3 = (True,'raman','taman',2,'shyam',89101)
i = 0
while(i < len(tuple3)):
    print(tuple3[i], end=" ")
    i+=1
```

### Other Operations on Tuples 

```python
# Join two tables
m = (1,2,3,4)
n = ('x','y','z')

l = m + n
print(type(l))
print(l)
```

```python
# Multiply two tables
m = (1,2,3,4)
p = m * 2
print(p)
```

```python
# count() - returns the number of times a specified value appears in the tuple.
letters = ('a','b','c','d','e','f','g','b','f','b')
print(letters.count('b'))
```

```python
# index() - Searches the tuple for a specified value and returns the position of where it was found 
# finds the first occurence of the specified value
# raises an exception if the value if not found
print(letters.index('f'))
```

# 2. Dictionary

Dictionaries are used to store data values in key:value pairs. A dictionary is a collection which is ordered, changeable and do not allow duplicates.
Dictionary items can be referred using the key name.

Dictionaries are changeable, meaning that we can change, add or remove items after the dictionary has been created.
Duplicates are not allowed in dictionaries.


```python
myDict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964,
    "year": 2020
}

print(type(myDict))
print(len(myDict))
print(myDict)
print(myDict["model"])
```

### get() method

```python
x = myDict.get("brand")
print(x)
```

### keys() method

```python
y = myDict.keys()
print(y)
```

```python
dict1 = { 
    "name": "akash",
    "age": 20,
    "marks": 91,
    "city": "ghaziabad"
}

print(dict1)

# updating the key value
dict1["marks"] = 94
print(dict1)
```

### update() method

```python
dict1.update({"city":"patiala"})
print(dict1)
```

```python
# adding an item in dictionary is done by using a new key index and assigning value to it.
# adding can also be done with update() method
dict1["rollno"] = 8
print(dict1)
```

### pop() method

The pop( ) method removes the item with the specified key name

```python
# pop(key_name)
# dict1.pop("rollno")
dict1.update({"hobby":"basketball"})
dict1
```

```python
dict1.pop("hobby")
dict1
```

```python
# popitem() - method removes the last inserted item
dict1.popitem()
dict1
```

```python
# del keyword - can delete the dictionary completely
del dict1
# print(dict1) -> gives an error because we have deleted the table
# clear() - also empties the dictionary
```

### Looping in Dictionaries

When looping through a dictionary, the return value are the keys of the dictionary, but there are methods to return the values as well.

```python
dict1 = { 
    "busNo": 23,
    "name": "punjab roadways",
    "from": "chandigarh",
    "to": "nainital"
}
```

```python
# Printing all key names in the dictionary, one by one
for i in dict1: 
    print(i)
```

```python
# Printing all values in dictionary
for x in dict1:
    print(dict1[x])
    
```

```python
# we can also use values() method to return values of a dictionary
for x in dict1.values():
    print(x)
```

```python
# use the keys() method to return the keys of a dictionary
for x in dict1.keys():
    print(x)
```

### items() method in Loops

```python
# Loop through both keys and values, by using the items() method
for x,y in dict1.items():
    print(x, y)
```

### copy() method


```python
dict2 = dict1.copy()
print(dict2)
```

```python
# dictionary can also be copied using dict() function
dict3 = dict(dict1)
dict3
```

## Nested Dictionaries

A dictionary can contain dictionaries, this is called nested dictionaries.

```python
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}

print(type(myfamily))
print(myfamily)
```

```python
# example2
child1 = {
  "name" : "Emil",
  "year" : 2004
}
child2 = {
  "name" : "Tobias",
  "year" : 2007
}
child3 = {
  "name" : "Linus",
  "year" : 2011
}

myfamily = {
  "child1" : child1,
  "child2" : child2,
  "child3" : child3
}

print(myfamily)
```
