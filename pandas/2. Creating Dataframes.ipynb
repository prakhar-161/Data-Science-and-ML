# Different Ways of Creating Dataframes

1. Using CSV
2. Using Excel
3. From python dictionary
4. From list of tuples
5. From list of dictionaries

```python
import pandas as pd
```

## 1. Using CSV

```python
df = pd.read_csv('weather.csv')
df
```

## 2. Using Excel

```python
df1 = pd.read_excel('weather.xls')
df1
```

## 3. From Python Dictionary

```python
customer_data = {
    'Id': ['1020161','1020162','1020163','1020164','1020165'],
    'Name': ['Kevin','Drake','Angela','Lisa','Liam'],
    'Country': ['USA','Columbia','India','England','Nigeria'],
    'Gender': ['M','M','F','F','M']
}

df2 = pd.DataFrame(customer_data)
print(df2)
```

```python
df2.set_index('Id', inplace=True)
```

```python
df2
```

```python
df2.loc['1020163']
```

## 4. From List of Tuples

```python
customer_val = [
    ('001','Prakhar','India','M'),
    ('002','Lisa','UK','F'),
    ('003','Nanda','India','F'),
    ('004','Matt','USA','M'),
    ('005','Zlatan','Sweden','M')
]

# mention the names of columns
# By default, columns will be given numbers 0,1,2,3,...
df3 = pd.DataFrame(customer_val, columns=["Id","Name","Country","Gender"])
df3
```

## From List of Dictionaries

```python
weather_data = [
    {'day':'1/1/2020', 'temperature': 10, 'windspeed': 10, 'event': 'Windy'},
    {'day':'1/2/2020', 'temperature': 15, 'windspeed': 6, 'event': 'Rain'},
    {'day':'1/3/2020', 'temperature': 34, 'windspeed': 4, 'event': 'Sunny'}
]

df4 = pd.DataFrame(weather_data)
df4
```
