# Sorting 

.sort_values(column_name) - sorts the values in the column in ascending order.

.sort_values(column_name, ascending=False) - sorts the values in the column in descending order.

```python
df.sort_values('SepalWidthCm')
```

```python
df.sort_values('SepalLengthCm', ascending=False)
```

* We can sort by multiple variables by passing a list of column names to sort_values.

```python
df.sort_values(['PetalWidthCm','PetalLengthCm'])
```

```python
# We can also specify directions
# first item in list is preferred
df.sort_values(['PetalLengthCm','PetalWidthCm'], ascending=[True,False])
```

# Subsetting

## Subsetting Columns 

To get the data of only some columns, we perform subsetting of columns.

To select multiple columns, mention their names inside two pairs of square brackets.

```python
df['SepalLengthCm']
```

```python
df[['SepalLengthCm','SepalWidthCm']]
```

```python
# other method
sepal_props = ['SepalLengthCm','SepalWidthCm']
df[sepal_props]
```

## Subsetting Rows

A common way to do this is by creating a logical condition to filter against.

### Subsetting based on numerical values 

```python
df['SepalLengthCm'] > 5.2
# returns boolean values
```

```python
df[df['SepalLengthCm'] > 5.2]
# we will get the rows for in which the SepalLengthCm > 5.2
```

### Subsetting based on Text

```python
df[df['Species'] == 'Iris-versicolor'].head()
```

### Subsetting based on multiple conditions 

```python
is_length = df['SepalLengthCm'] == 5.6
is_species = df['Species'] == 'Iris-versicolor'

df[is_length & is_species]
```

```python
# in one line
df[(df['SepalLengthCm'] == 6.1) & (df['Species'] == 'Iris-versicolor')]
```

### Subsetting using .isin( )

To filter on multiple values of a categorical variable(column), the easiest way is to use isin() method.

This takes in a list of values to filter for.

```python
# here, filtering all the flower details on basis of 'PetalLengthCm' whose values are 4.7 and 1.3
df[ df['PetalLengthCm'].isin([4.7,1.3]) ]
```
