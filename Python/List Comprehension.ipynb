# List Comprehension

List comprehension offers a shorter syntax when we want to create a new list based on the values of an existing list. It provides a concise way to create lists. 

Syntax: newlist = [expression for item in iterable if condition == True]

The condition is like a filter that only accepts the items that valuate to True.
The iterable can be any iterable object, like a list, tuple, set etc.
The expression is the current item in the iteration, but it is also the outcome, which you can manipulate before it ends up like a list item in the new list.

The return value is a new list, leaving the old list unchanged.

```python
# For eg - we have to add the cars whose names contain letter 'r' in them
# old method

cars = ['ferrari','bentley','bmw','audi','rolls royce']
newCarList = []

for x in cars:
    if 'r' in x:
        newCarList.append(x)
        
print(newCarList)
```
