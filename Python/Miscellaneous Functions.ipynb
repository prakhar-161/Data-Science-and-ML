# Map Functions in Python

map() function returns a map object(which is an iterator) of the results after applying the given function to each item of a given iterable (list, tuple etc.)

Syntax: 
map(function, iterable)

```python
def even_odd(num):
    if num%2==0:
        return "The number {} is even".format(num)
    else:
        return "The number {} is odd".format(num)
```

```python
lst = [1,2,3,4,5]

print(list(map(even_odd, lst)))
```

# Lamba Functions in Python 

A lambda function is a small anonymous function a function with no name).
A lambda function can take any number of arguments, but can only have one expression.

Syntax: lambda arguments : expression

If we have a single expression inside a function, then we can make use of lambda function.

```python
x = lambda a : a + 5
print(x(3))
```

```python
x = lambda a, b : a * b
print(x(7, 4))
```

```python
x = lambda a, b, c : a - b + c
print(x(2, 4, 9))
```

```python
x = lambda n : n%2==0
print(x(38))
print(x(49))
print(x(0))
```

# Filter Function in Python

```python
def even(num):
    if num%2==0:
        return True
```

```python
lst=[1,2,3,4,5,6,7,8,9,0]
```

```python
list(filter(even, lst))
```

```python
# Using lambda function
list(filter(lambda num:num%2==0, lst))
```

```python
# also map() function
list(map(lambda num:num%2==0, lst))
```
