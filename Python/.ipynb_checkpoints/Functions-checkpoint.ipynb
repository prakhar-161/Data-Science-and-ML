# Functions in Python

A function is a block of code which only runs when it is called.
You can pass data, known as parameters, into a function.
A function can return data as a result.

A function is defined using the def keyword.

```python
def even_odd(num):
    if num%2==0:
        print("number is even")
    else:
        print("number is odd")
```

```python
even_odd(208)
```

```python
def greet():
    print("Good morning")
    
greet()
```

```python
# function returning a value which can be an int, float, string or bool.
def say_hello():
    return "Hello, my fellow mate"

val=say_hello()
print(val)
```

```python
def add(n1,n2):
    return n1+n2

sum=add(20,32)
print(sum)
```

### Positional and Keyword Arguments

```python
# In the below example 
# first arg is 'positional argument'
# second arg is 'keyword argument'
def hello(name,age=19):
    print("My name is {} and age is {}".format(name,age))
    
hello('Prakhar')
```

```python
# If we externally specify value of keyword argument, then that value is there in the output.
hello('yoyo',21)
```

### Arbitrary Arguments (*args)

```python
def myfunc(*cars):
    print("My favourite car is " + cars[1])
    
myfunc('ferrari','aston martin','bentley','lamborghini')
```

### Keyword Arguments

We can also send arguments with the key = value syntax.
This way the order of the arguments does not matter.

```python
def myfunc(car1,car2,car3):
    print("My favourite car is " + car3)
    
myfunc(car2='bentley',car3='ferrari',car1='rolls royce')
```

### Arbitrary Keyword Arguments (**kwargs)

If you do not know how many keyword arguments that will be passed into your function, add two asterisk: ** before the parameter name in the function definition.

This way the function will receive a dictionary of arguments, and can access the items accordingly.

```python
def func(*args,**kwargs):
    print(args)
    print(kwargs)
    
func("Prakhar","Dhama",age=19,year=2002)
```

```python
def myfunc(**student):
    print("This student's first name is " + student["fname"] + " and last name is " + student["lname"])
    
myfunc(fname="Prakhar", lname="Dhama")
```

### Default Parameter Value

If we call the function without argument, it uses the default value.

```python
def mycar(car="Bentley"):
    print("I have " + car)
    
mycar("Lamborghini")
mycar()
mycar("Supra")
```

### Passing a List as an Argument

You can send any data types of argument to a function (string, number, list, dictionary etc.), and it will be treated as the same data type inside the function.

```python
def myfood(food):
    for x in food:
        print(x)
        
fruits = ["apple","banana","mango","papaya"]
myfood(fruits)
```

### Pass Statement

The function definitions cannot be empty, but if you for some reason have a function definition with no content, put in the pass statement to avoid getting an error.

```python
def myfunc():
    pass
```

## Example

```python
lst = [1,2,3,4,5,6,7]
```

```python
def evenoddsum(lst):
    even_sum=0
    odd_sum=0
    for i in lst:
        if i%2==0:
            even_sum += i
        else:
            odd_sum += i
    return even_sum,odd_sum
```

```python
evenoddsum(lst)
```
