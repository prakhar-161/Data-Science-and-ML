# 2. Sets

An unordered collection data type that is iterable, not mutable (unchangeable), and has no duplicate elements. Python's set class represents the mathematical notation of a set. This is based on a data structure called hash table. Elements are stored inside curly brackets { }.

Set items are unchangeable, but you can remove items and add new items.
Set items cannot be referenced by any index or key.
Once a set is created, you cannot change its items, but you can remove items and add new items.

Sets cannot have two items with the same value.
A set can contain different data types.

Since, we can't access items in a set through any index or key, so, we use use FOR Loop or IN Keyword.

We can also join two sets through union and can perform intersection on sets too.


```python
# Defining an empty set

set_var=set()
print(set_var)
print(type(set_var))
```

```python
set_var = {"Jack", "Daniel", "Tom", "Banks"}
print(set_var)
print(type(set_var))
```

```python
print(len(set_var))
```

### Set() Constructor

```python
# set() - It is also possible to use the set() constructor to make a set.

set_ex=set(("apple", "mango", "pear", "banana"))
set1=set()
print(set_ex)
```

```python
for item in set_ex:
    print(item)
```

### 'IN' Keyword 

```python
# return boolean value i.e either true or false
print("mango" in set_ex)
```

```python
print("strawberry" in set_ex)
```

### add() method

```python
# add() - To add one item to a set use the add() method.
n = 5
i=10
while len(set1)!=n:
    set1.add(i)
    i+=1
    
print(set1)
```

```python
set_ex.add("kiwi")
print(set_ex)
```

### update() method

```python
# update()- To add items from another set into the current set, use the update() method.
fruits = {"pineapple", "cherry", "papaya"}
set_ex.update(fruits)
print(set_ex)
```

### remove(), discard(), pop(), clear() method

```python
# To remove an item in a set, use the remove(), or the discard() method.
# If the item to remove does not exist, remove() will raise an error.
# If the item to remove does not exist, discard() will NOT raise an error.
# pop() - return value of the pop() method is the removed item. Sets are unordered, so when using the pop() method, you do not know which item that gets removed.
# clear() - this method empties the set.

set2 = {"bmw", "audi", "jaguar", "honda", "maruti", "lexus", "toyota", "kia", "hyundai"}

set2.remove("maruti")
print(set2)
```

```python
set2.discard("tata")
print(set2)

set2.discard("kia")
print(set2)
```

```python
set2.pop()
print(set2)
```

```python
# in previous run, "hyundai" got deleted.
x = set2.pop()
print(x)
print(set2)
```

```python
set2.clear()
print(set2)
# empties the set
# Also, del keyword deleted the set completely
```

### union(), update(), intersection_update(), intersection() 

```python
set_a = {1,2,3}
set_b = {'A','B','C'}

# union() - returns a new set containing all items from both sets
set_c = set_a.union(set_b)
print(set_c)
print(set_b)
```

```python
# update() - method that inserts all the items from one set into another 'randomly'.
set_a.update(set_b)
print(set_a)
```

```python
# Both union() and update() will exclude any duplicate items.

# intersection_update() - this method will keep only the items that are present in both sets.
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.intersection_update(y)
print(x)
```

```python
# intersection() - method will return a new set, that only contains the items that are present in both sets.
u = {"car","bus","truck","bicycle","train","airplane"}
v = {"train","truck"}
w = u.intersection(v)
print(w)
```
