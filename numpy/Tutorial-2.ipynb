# Numpy Functions 

```python
import numpy as np
```

### 1. Zeros

```python
arr1=np.zeros(5)
arr2=np.zeros((2,3))
print('1-D : ',arr1)
print('2-D : ',arr2)
```

### 2. Ones

```python
a1=np.ones(4)
a2=np.ones((2,2))
print(a1)
print(a2)
```

### 3. Eye 

```python
# This function creates an array with all the diagonal elements as 1 and rest as 0(in square matrix),
# In case of a non-square matrix, the values are still 1 for diagonal(upto  which it can draw diagonal) and rest are zero.

a3=np.eye(3)
a4=np.eye(3,4)
print(a3)
print(a4)
```

### 4. Diag

```python
# This funtion creates a 2-dimensional array with all the diagonal elements as the given value and rest as 0(in a square matrix).

a5=np.diag([1,2,3,4])
np.diag(a5)
# print(a5)
```

### 5. Randint

```python
# This function is used to generate random number between a given range.
# rand(min,max,total_no_of_values)

a6=np.random.randint(1,11,2)
a6
```

### 6. Rand

```python
# This function is used to generate random values between 0 to 1
# rand(no_of_values)

a7=np.random.rand(4)
a8=np.random.rand(2,2)
print(a7)
print(a8)
```

### 7. Randn

```python
# This function is used to generate random values close to 0. This may return positive and negative numbers.
# rand(no_of_values)

a9=np.random.randn(3)
a9
```
