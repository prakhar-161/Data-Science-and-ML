# Working on 2-D Arrays

```python
import numpy as np
```

## Indexing in 2-D Arrays

```python
matrix=[]
row=int(input("Enter no of rows: "))
col=int(input("Enter no of columns: "))
for i in range(row):
    a=[]
    for j in range(col):
        val=int(input("Enter value: "))
        a.append(val)
    matrix.append(a)
# converting the matrix into 2-D array
arr=np.array(matrix)
for i in range(row):
    for j in range(col):
        print(arr[i,j], end=" ")
    print()
print(type(arr))
```

## Slicing 

```python
a = np.array([[10,20,30],[40,50,60],[70,80,90]])
print(a)

# a[row range, col range]
print(a[:2])     # returns row from 0 to 1 and since col is not given then all cols are printed
print(a[:2,1:3])
```

```python
np.random.seed(100)
a=np.random.randint(1,500,24).reshape(4,6)
print(a)

print(a[1:,3:])
```
