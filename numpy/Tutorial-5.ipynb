```python
import numpy as np
```

# Some Functions


## 1. Arange

```python
# returns a list of numbers from start to end.
A1 = np.arange(100)
A2 = np.arange(20,100,10)
print(A1)
print(A2)
```

## 2. Range


```python
print(range(0,10))
```

```python
print(list(range(0,10)))
```

```python
print(tuple(range(0,20,2)))
```

## 3. Random

```python
# random.permutation
A = np.random.permutation(np.arange(10))
print(A)
```

```python
# random.randint
np.random.randint(10,20)
```

```python
A = np.random.rand(10)
print(A)
```

```python
B = np.random.rand(2,4)
print(B)
```

## 4. Reshape

```python
# reshape array
C = np.arange(100).reshape(4,25)
print(C.shape)
```
