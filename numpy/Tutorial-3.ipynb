# More Working on Arrays

```python
import numpy as np
```

## Indexing in 1-D Arrays

```python
a=[]
size=int(input("Enter size : "))
for i in range(size):
    num=int(input("Enter number : "))
    a.append(num)
arr=np.array(a)
for i in range(arr.size):
    print(arr[i])

sum=0
for i in range(arr.size):
    sum=sum+arr[i]
print('sum : ',sum)
```

## Slicing in 1-D Arrays

```python
# arr_name[start:end:step]
# arr_name[start:end+1]
# arr_name[start: ]
# arr_name[ :end+1]
# arr_name[ : ]
```

```python
a1=[]
size=int(input("Enter size : "))
for i in range(size):
    num1=int(input("Enter number : "))
    a1.append(num1)
arr1=np.array(a1)
for i in range(arr1.size):
    print(arr1[i])

a2=arr1[1::2]
print(a2)
```

## Reshaping of Arrays

```python
arr=np.random.randint(1,50,12)
print(arr.shape)

arr1=arr.reshape(3,4)
print(arr1)
print(arr1.shape)

arr2=arr.reshape(4,3)
print(arr2)

arr3=arr.reshape(2,6)
print(arr3)

# when column or row is unknown
arr4=arr.reshape(4,-1)
print(arr4)
```
